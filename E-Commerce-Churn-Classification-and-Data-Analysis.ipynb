## Business Problem Understanding

**Context**

E-commerce company is experiencing churn, which can be referred to as customers who no longer use the E-commerce platform. Churn is considered a loss for the company, as the longer customers stay away from the E-commerce platform, the fewer they become. Therefore, the company wants to identify which customers will remain loyal and which ones will leave the E-commerce platform. This information can be utilized by the company to take marketing steps that align with the characteristics of customers.

Target :
<br>
0 : Customers who remain loyal and continue to shop on the E-commerce platform
<br>
1 : Customers who have stopped using the E-commerce platform

**Problem Statement**

Customers who have the potential to churn can reduce the company's revenue. Therefore, there is a need for action from the company to increase revenue. If the company targets marketing initiatives such as promotions for all customers without first filtering, it will incur significant expenses. The company needs to be more efficient in determining its expenditures by identifying which customers are potentially going to churn.

**Goals**

Therefore, based on this issue, the company aims to have the ability to predict the likelihood of a customer churning or not. This way, promotions can be more targeted towards customers who are likely to churn.

The company also wants to understand the factors that contribute to customer churn or retention. This knowledge will enable the company to create a more effective plan for promotions or other strategic initiatives.

**Analytics Approach**

We will analyze the data to identify patterns that differentiate customers who are potentially going to churn or not. Subsequently, we will build a classification model that will assist the company in detecting whether a customer is likely to churn or not.

**Metric Evaluation**



**Type 1 error: False Positive**
<br>
Consequence: Wasted promotional costs incurred by the company

**Type 2 error: False Negative**
<br>
Consequence: The company loses a customer

Based on the consequences, we will create a model that can reduce (False Positives/True Positives)?

# Categorical Data Checking

```python
# Print each unique categorical data to check any typo or any same type category with different names.

unique_dict = {}

for col in df.columns:                                                  # Search for unique categorical data by looping and save it in a dictionary
    if df[col].dtype == 'object' or df[col].dtype == 'category':
        unique_dict[col] = df[col].unique()

for col, unique in unique_dict.items():                                 # Print each unique categorical data by looping
    print(f"'{col}' unique:\n{unique}\n")
```

There is not any typo but there is some category of same type but with different names. We need to replace it's name and make it to one category type.

```python
# Replace any category of same type into one category type

# PreferredLoginDevice Category
df['PreferredLoginDevice'].replace(['Phone'], ['Mobile Phone'], inplace=True)

# PreferredPaymentMode Category
df['PreferredPaymentMode'].replace(['CC'],['Credit Card'],inplace=True)
df['PreferredPaymentMode'].replace(['Cash on Delivery'],['COD'],inplace=True)

# PreferedOrderCat Category
df['PreferedOrderCat'].replace(['Mobile'],['Mobile Phone'],inplace=True)
```

# Outliers

```python
def numerical_boxplots(data):                                                           # Define function to create boxplot for each numerical data
    numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
    num_plots = len(numerical_columns)
    num_rows = (num_plots + 2) // 3

    fig, axes = plt.subplots(num_rows, 3, figsize=(18, 6 * num_rows))
    axes = axes.ravel()                                                                 # Flatten the 2D array of axes

    for i, column in enumerate(numerical_columns):                                      # Create boxplot for each numerical data by looping
        sns.boxplot(data=data, x=column, ax=axes[i])
        axes[i].set_title(f'Box Plot of {column}')

    # Hide any empty subplots
    for i in range(num_plots, num_rows * 3):
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()


numerical_boxplots(df)
```

From each boxplot above, there isn't any necesarry to remove outliers. This beacuse all the values contain valuable information that doesn't affect classification negatively.
